```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
wiki_url = "https://en.wikipedia.org/wiki/List_of_African_countries_by_population"
table_class = "sortable wikitable static-row-numbers jquery-tablesorter"
```

```python
response=requests.get(wiki_url)
print(response.status_code)
```

```python
soup=BeautifulSoup(response.text, 'html.parser')
table_1 = soup.find('table', {'class': "wikitable"})
```

```python
table_1
```

```python
df = pd.read_html(str(table_1))
```

```python
df
```

```python
print(df)
```

```python
African_countries = pd.DataFrame(df)
```

```python
African_countries.to_csv("African_countries_population.csv")
```
